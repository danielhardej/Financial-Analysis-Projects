```python

from io import BytesIO
import requests
import time
import pandas as pd
import os

BASE_URL = "https://www.alphavantage.co/query?"
API_KEY = os.environ.get("ALPHA_VANTAGE_API_KEY") 
```

```python
def earnings_calendar_api(api_key, horizon, symbol=None):
    if symbol is not None:
        url = f'{BASE_URL}function=EARNINGS_CALENDAR&symbol={symbol}&horizon={horizon}&apikey={api_key}'
        response = requests.get(url)
    else:
        url = f"{BASE_URL}function=EARNINGS_CALENDAR&horizon={horizon}&apikey={api_key}"
        response = requests.get(url)

    return pd.read_csv(BytesIO(response.content))

horizon=["3month","6month","12month"]
earnings_calendar_api(API_KEY,horizon[0])
```

```python
from datetime import datetime

start = datetime(2020, 1, 1)
end = datetime(2023, 4, 4)
```

```python
def to_unixtime(dt):
    epoch = datetime.utcfromtimestamp(0)
    return int((dt - epoch).total_seconds())
```

```python
aapl_query_url = 'https://query1.finance.yahoo.com/v7/finance/download/aapl?period1={}&period2={}'.format(to_unixtime(start), to_unixtime(end))
nasdaq = pd.read_csv(aapl_query_url, index_col=[0],parse_dates=False)
```
